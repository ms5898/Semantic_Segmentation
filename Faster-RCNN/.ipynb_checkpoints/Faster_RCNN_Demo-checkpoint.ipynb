## Only keep 0.7 score_thresh and do NMS to every class with nms_thresh = 0.3

```python
score_thresh = 0.7
nms_thresh = 0.3
scores = output[0]['scores'].data.numpy()
index = np.where(scores > score_thresh)[0].tolist()

scores_ts = output[0]['scores'][index]
boxes_ts = output[0]['boxes'][index]
labels_np = output[0]['labels'].data.numpy()[index]
labels_ts = output[0]['labels'][index]
```

```python
scores_keep = []
boxes_keep = []
labels_keep = []
for key in category_map.keys():
    index = np.where(labels_np==key)[0].tolist()
    if len(index) != 0:
        scores = scores_ts[index]
        boxes = boxes_ts[index]
        labels = labels_ts[index]
        keep = torchvision.ops.nms(boxes, scores, nms_thresh).data.numpy().tolist()
        for i in keep:
            scores_keep.append(scores.data.numpy()[i])
            boxes_keep.append(boxes.data.numpy()[i])
            labels_keep.append(labels.data.numpy()[i])
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1)
for i in range(len(labels_keep)):
    bbox = boxes_keep[i]
    category_id = labels_keep[i]
    rect = plt.Rectangle((bbox[0],bbox[1]), bbox[2]-bbox[0], bbox[3]-bbox[1], fill=False, edgecolor = 'red',linewidth=1)
    plt.text(bbox[0], bbox[1], str(category_id) + category_map[category_id], fontsize=12, color = "r", bbox = dict(facecolor = "g", alpha = 0.2))  
    ax.add_patch(rect)
img_show = np.array(img)
plt.imshow(img_show)
plt.show()
```

# Step by Step

# Build a VGG16 Faster R-CNN

```python
import torchvision
from torchvision.models.detection import FasterRCNN
from torchvision.models.detection.rpn import AnchorGenerator

backbone = torchvision.models.vgg16(pretrained=True).features

backbone.out_channels = 512

anchor_generator = AnchorGenerator(sizes=((64,128,256),),
                                   aspect_ratios=((0.5, 1.0, 2.0),))

# put the pieces together inside a FasterRCNN model
model2 = FasterRCNN(backbone,num_classes=91,
                   min_size=600, max_size=1000,
                   rpn_anchor_generator=anchor_generator)
```

# Scale the input image before given to the network

* reference: https://github.com/pytorch/vision/blob/master/torchvision/models/detection/faster_rcnn.py

```python
features = []
def hook(module, input, output):
    features.append(input)
    features.append(output)
    
for name, m in model2.named_modules():
    if name == 'backbone':
        m.register_forward_hook(hook)

model2.eval()
output =model2(image)
```

```python
print('size of original image: {}'.format(image[0].size()))
print('size of scaled image: {}'.format(features[0][0].size()))
print('The min size for scaled image: {}'.format(model2.transform.min_size))
print('The max size for scaled image: {}'.format(model2.transform.max_size))
```
